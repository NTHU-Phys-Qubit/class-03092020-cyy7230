```python
import matplotlib.pyplot as plt
import numpy as np
```

## Problem 1

Plot sin(x)

```python
plt.figure(figsize=(6, 2))
t = 6*np.pi
x = np.linspace(-t, t, 256, endpoint=True)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.xlim(-10.0, 10.0)
plt.ylim(-1.0, 1.0)

plt.xticks(np.linspace(-10, 10, 21, endpoint=True))
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
plt.show()
```
